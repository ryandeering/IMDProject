# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Preprocessing

```python
fileName = 'student-mat.csv'

data = pd.read_csv(fileName)

data.head()
```

```python
fileName = 'student-por.csv'

data2 = pd.read_csv(fileName)

data2.head()
```

```python
print("Maths Students Missing Data: \n", data.isnull().sum())
```

```python
print("Portugese Students Missing Data: \n", data2.isnull().sum())
```

```python
data.info()
```

```python
#No nulls recorded. We're going to merge the two datasets based on their classes. 

data.insert(0, 'class', 'Maths')
data.head()

data2.insert(0, 'class', 'Portugese')
data2.head()
```

Cool. Let's do a merge.

```python
student_merge = pd.concat([data, data2], axis=0)
```

```python
student_merge
```

```python
student_merge.reset_index(drop=True, inplace=True)
print(student_merge)
duplicate = student_merge[student_merge.duplicated()] 
print(duplicate)
#No duplicate rows.
```

# Minor EDA

```python
namesCol = list(student_merge.columns)


correlations = student_merge.corr()
print("Correlation Matrix")
# plot correlation matrix
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(namesCol)
ax.set_yticklabels(namesCol)
plt.show()
print("Correlation Table")
# Correlation Table, note this does not export easily
corr = student_merge.corr()
corr.style.background_gradient().set_precision(2)
```
